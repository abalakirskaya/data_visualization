```python
import pandas as pd
```

```python
import altair as alt
```

```python
df = pd.read_csv('/Users/alisa/Desktop/UCU/data_visualization/HW1/GLB.Ts_dSST.csv', header=1 )
df.head()
```

```python
df.drop(['J-D', 'D-N', 'DJF', 'MAM', 'JJA', 'SON'], axis = 1, inplace = True)
```

```python
df.set_index('Year', inplace=True)
col = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
for i in df.index:
    for j in col:
        df[j][i] = float(df[j][i])
```

```python
df['mean'] = df.mean(axis=1)
```

```python
df = df.reset_index()
```

```python
chart = alt.Chart(df).mark_point(filled = True, color = "#ffa500").encode(
    x = alt.X(field='Year', type = 'quantitative', title = 'Year', scale=alt.Scale(domain=[1880, 2017])),
    y = alt.Y(field = 'mean', title = 'Degree, Â°', type = 'quantitative'),
    tooltip=['Year', 'mean']
)
#.properties(width = 950, height = 500, background = '#F9F9F9', padding = 25)
```

```python
line = line = pd.DataFrame({
    'Year': [1880, 2017],
    'Degree':  [0, 0],
})

line_plot = alt.Chart(line).mark_line(color= 'grey').encode(
    x= 'Year',
    y= 'Degree',
)
```

```python
alt.layer(line_plot, chart).properties(width = 950, height = 500, background = '#F9F9F9', padding = 25, title = 'Annual Global Surface Temperature Relative to Late 19th Century Average').configure_title(anchor='start')

```

```python
text = chart.mark_text(
    align='left',
    baseline='middle',
    fontSize=10,
    dx=7
).encode(
    text=alt.Text('Year')
)
```

```python
alt.layer(line_plot, chart, text).properties(width = 950, height = 500, background = '#F9F9F9', padding = 25, title = 'Annual Global Surface Temperature Relative to Late 19th Century Average').configure_title(anchor='start')
```
